```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_iris
```

```python
data = pd.read_csv('migration.csv')
data
```

```python

# Set the 'Country Name' as the index
data.set_index('Country Name', inplace=True)
```

```python
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)
```

```python
data
```

```python
# Perform PCA
pca = PCA(n_components=2)  # Reduce to 2 principal components
principal_components = pca.fit_transform(data_scaled)
```

```python
pca_df = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2'], index=data.index)

# Plot the results
plt.figure(figsize=(10, 8))
plt.scatter(pca_df['PC1'], pca_df['PC2'], alpha=0.5)
plt.title('PCA of Migration Data')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')

# Annotate the points with country names
for i, txt in enumerate(pca_df.index):
    plt.annotate(txt, (pca_df['PC1'][i], pca_df['PC2'][i]), fontsize=8)

plt.show()

# Explained variance ratio
print("Explained variance ratio:", pca.explained_variance_ratio_)

```
